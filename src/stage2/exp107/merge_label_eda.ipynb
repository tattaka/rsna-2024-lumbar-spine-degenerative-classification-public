```python
import pandas as pd
import numpy as np

axial_t2_df = pd.read_csv(f"../../../input/Axial_T2.csv")
sagittal_t1_df = pd.read_csv(f"../../../input/Sagittal_T1.csv")
sagittal_t2_df = pd.read_csv(f"../../../input/Sagittal_T2-STIR.csv")
```

```python
train_coord_df = sagittal_t1_df.drop(
    columns=["x_l", "x_r", "y_l", "y_r"], inplace=False
).merge(
    sagittal_t2_df.loc[
        :, ["study_id", "series_id", "level", "label", "x", "y", "instance_number"]
    ],
    on=["study_id", "level"],
)
train_coord_df = train_coord_df.merge(
    axial_t2_df.loc[
        :,
        [
            "study_id",
            "series_id",
            "level",
            "label_l",
            "label_r",
            "x_l",
            "x_r",
            "y_l",
            "y_r",
            "instance_number_l",
            "instance_number_r",
        ],
    ].rename(
        columns={
            "series_id": "series_id_z",
            "label_l": "label_l_ax",
            "label_r": "label_r_ax",
            "x_l": "x_l_ax",
            "y_l": "y_l_ax",
            "x_r": "x_r_ax",
            "y_r": "y_r_ax",
            "instance_number_l": "instance_number_l_ax",
            "instance_number_r": "instance_number_r_ax",
        }
    ),
    on=["study_id", "level"],
)

train_coord_df = train_coord_df[
    ~(
        train_coord_df.label_l.isna()
        | train_coord_df.label_r.isna()
        | train_coord_df.label.isna()
        | train_coord_df.label_l_ax.isna()
        | train_coord_df.label_r_ax.isna()
    )
]
train_coord_df
```
